```python
# Collections 
# Lists

# Ordered, Mutable, Allow Duplicate Values

myList = [10,20,30,40,50]

print(myList)
print(myList[3]) # access based on index

myList.append('Hema') # append add element at last index
print(myList)

myList.insert(2,25)
print(myList)

print(myList.pop()) # return the last element and remove the last value
print(myList)

print(f"The Length of the list is : {len(myList)}")

print("Maximum Value :", max(myList))
print("Minimum Value :", min(myList))
print("Total Sum :", sum(myList))

# Remove the value by index
myList.remove(25)
print(myList)

# Check membership
print(20 in myList)

mlist =[10,20,30,40,50]
print(myList is mlist)

print(myList[1:4])

myList.extend([1,2,3,4,5]) # adding element at last
print(myList)

# Delete using index number
del myList[1]
print(myList)

myList.clear()
print(myList)
```

```python
# Loops in LISTs

myData = [1,2,3,4,5,6,7,8]

for num in myData:
    print(num, end = " ")
print()    

i = 0
while i < len(myData):
    print(myData[i] , end = " ")
    i += 1
print()    

myData.reverse()
print(myData)

myData.sort()
print(myData)

myData.sort(reverse = True)
print(myData)

# Copy the List

copy_number = myData.copy()
print(copy_number)

# Sort Way to Create List

squares = [x ** 2 for x in range(1,6)]
print(squares)
```

```python
# As Tuple is immutable, then for modifying the tuple 
# We have to convert tuple into list 
# Then Modify the List
# Lastly, Convert the list into tuple

number_tuple = (23, 'Hema', 78.33)
print(number_tuple)

number_tuple = list(number_tuple)
print(number_tuple)

number_tuple.append('Chandu')
print(number_tuple)

number_tuple = tuple(number_tuple)
print(number_tuple)
```

```python
# Set -- Unordered , Unique Elements, Mutable, Supports Set Operations(Union, Intersection)

names = { "alex", "bob", "john", "catty"}
print(names)

for val in names:
    print(val)

names.add('krishna')
print(names)

# To add many elements
names.update(['rohini', 'pawani'])
print(names)

# Removing the value from set
names.remove('rohini')
print(names)


names.discard('paaru')

names.pop();
print(names)
```

```python
# Set Operations

A = {1,2,3,4,5}
B = {4,5,6,7,8}

# Union
print(A|B)
print(A.union(B))

# Intersection
print(A & B)
print(A.intersection(B))

# Difference
print(A - B)
print(A.difference(B))

# Symmetric Difference
print(A ^ B)
print(A.symmetric_difference(B))
```

```python
# Movie Ticket Booking System
#Task2
movies = []
theaters = set()
ticket_price = (200, 300, 400)
customers = {}

def add_movie(title, genre, showtime):
    movies.append({'title': title, 'genre': genre, 'showtime': showtime})
    print(f"Added Movie -------------------------------------------\nTitle: {title}\nGenre: {genre}\nShowtime: {showtime}\n")

def remove_movie(title):
    for movie in movies:
        if movie['title'] == title:
            movies.remove(movie)
            print(f"Removed Movie --------------------------------------\nTitle: {title}\nGenre: {movie['genre']}\nShowtime: {movie['showtime']}\n")

def add_theater(theater_name):
    theaters.add(theater_name)
    print("Theater Added:", theater_name)

def add_customer(customer_name):
    customers[customer_name] = []
    
def book_ticket(customer_name, movie_title):
    for movie in movies:
        if movie["title"] == movie_title:
            customers[customer_name].append(movie_title)
            print(f"Ticket booked for '{customer_name}' to watch '{movie_title}'.")
            return

def print_data():
    print("\nMovies Available:")
    for movie in movies:
        print(f"  {movie['title']} ({movie['genre']}) at {movie['showtime']}")

    print("\nTheaters Available:")
    for theater in theaters:
        print(f"  {theater}")

    print("\nTicket Pricing:")
    print(f"  Prices: {ticket_price}")

    print("\nCustomer Bookings:")
    for customer, bookings in customers.items():
        print(f"  {customer}: {', '.join(bookings) if bookings else 'No bookings yet'}")

add_movie("Avengers: Endgame", "Action", "18:00")
add_movie("Toy Story 4", "Animation", "14:00")
add_movie("Joker", "Drama", "20:00")

# remove_movie("Toy Story 4")

add_theater("Cinepolis")
add_theater("Inox")
add_theater("PVR Cinemas")

print("---------------------------------------------------------------------------------------------")

add_customer("Alice")
add_customer("Bob")
add_customer("Charlie")

book_ticket("Alice", "Avengers: Endgame")
book_ticket("Bob", "Joker")
book_ticket("Charlie", "Toy Story 4")

print("---------------------------------------------------------------------------------------------")

print_data()
```
