```python
# Movie Ticket Booking System
#Task2
movies = []
theaters = set()
ticket_price = (200, 300, 400)
customers = {}

def add_movie(title, genre, showtime):
    movies.append({'title': title, 'genre': genre, 'showtime': showtime})
    print(f"Added Movie -------------------------------------------\nTitle: {title}\nGenre: {genre}\nShowtime: {showtime}\n")

def remove_movie(title):
    for movie in movies:
        if movie['title'] == title:
            movies.remove(movie)
            print(f"Removed Movie --------------------------------------\nTitle: {title}\nGenre: {movie['genre']}\nShowtime: {movie['showtime']}\n")

def add_theater(theater_name):
    theaters.add(theater_name)
    print("Theater Added:", theater_name)

def add_customer(customer_name):
    customers[customer_name] = []
    
def book_ticket(customer_name, movie_title):
    for movie in movies:
        if movie["title"] == movie_title:
            customers[customer_name].append(movie_title)
            print(f"Ticket booked for '{customer_name}' to watch '{movie_title}'.")
            return

def print_data():
    print("\nMovies Available:")
    for movie in movies:
        print(f"  {movie['title']} ({movie['genre']}) at {movie['showtime']}")

    print("\nTheaters Available:")
    for theater in theaters:
        print(f"  {theater}")

    print("\nTicket Pricing:")
    print(f"  Prices: {ticket_price}")

    print("\nCustomer Bookings:")
    for customer, bookings in customers.items():
        print(f"  {customer}: {', '.join(bookings) if bookings else 'No bookings yet'}")

add_movie("Avengers: Endgame", "Action", "18:00")
add_movie("Toy Story 4", "Animation", "14:00")
add_movie("Joker", "Drama", "20:00")

# remove_movie("Toy Story 4")

add_theater("Cinepolis")
add_theater("Inox")
add_theater("PVR Cinemas")

print("---------------------------------------------------------------------------------------------")

add_customer("Alice")
add_customer("Bob")
add_customer("Charlie")

book_ticket("Alice", "Avengers: Endgame")
book_ticket("Bob", "Joker")
book_ticket("Charlie", "Toy Story 4")

print("---------------------------------------------------------------------------------------------")

print_data()
```
