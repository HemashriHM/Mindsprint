```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Displaying the dataset:
df = pd.read_csv('employee_data.csv')
df.head()
```

```python
# Average Salary in each department:
avg_salary = df.groupby('Department')['Salary'].mean()
print(avg_salary)
```

```python
# Highest salary Employee in the company
max_salary = df['Salary'].max()
print(f"Highest Salary is: {max_salary}")

employee_details = df[df['Salary'] == max_salary]
print(employee_details)
```

```python
# Lowest salary Employee in the company
min_salary = df['Salary'].min()
print(f"Lowest Salary is: {min_salary}")

employee_details = df[df['Salary'] == min_salary]
print(employee_details)
```

```python
# Count of number of employees in each department
employee_count = df.groupby('Department')['Employee_ID'].count()
print(employee_count)
```

```python
# Visualization of the above analysis
employee_count.plot(kind = 'bar', color = 'c')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.title('Number of Employees in Each Department')
plt.show()
```

```python
# Salary distribution visual representation using pie chart
salary_distribution = df.groupby('Department')['Salary'].sum()
plt.figure(figsize=(8,8))
salary_distribution.plot(kind='pie', autopct='%1.1f%%', shadow = 'True')
plt.title('Salary Distribution by Department')
plt.ylabel('')
plt.show()
```

```python
# Scatter plot showing salary vs experience 
plt.figure(figsize=(8,4))
plt.scatter(df['Experience'], df['Salary'])
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Scatter Plot of Salary vs Experience')
plt.grid(True)
plt.show()
```

```python
# Employees earning more than average salary
average_salary = df['Salary'].mean()
employees_above_average = df[df['Salary'] > average_salary]
print(employees_above_average)
```

```python
# Sorting employees based on salary in descending order
sorted_employees = df.sort_values(by='Salary', ascending=False)

print(sorted_employees)
```
