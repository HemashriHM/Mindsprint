```python
import numpy as np
data = np.array([10,12,23,43,12,32,100])

Q1 = np.percentile(data, 10)
Q3 = np.percentile(data, 45)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = data[(data < lower_bound) | (data > upper_bound)]
print("Outliers: ", outliers)
```
