```python
x = [1,2,3,4,5]
y = [10,20,25,30,40]
plt.plot(x,y, marker='o', linestyle = '-', color= 'b')
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Lable")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
x=[1,2,3,4,5]
y=[10,20,25,30,40]
plt.scatter(x,y,marker='*',color='r')
plt.xlabel("X-axis")
plt.ylabel("Y-lable")
plt.title("Line Plot")
plt.grid()
plt.show()
```
